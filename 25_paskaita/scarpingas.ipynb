```python
import requests
from bs4 import BeautifulSoup

doc1 = requests.get("https://www.skelbimai.lt/skelbimai/nekilnojamasis-turtas/butai")

if doc1.status_code != 200:
    print('prisijungimas nepavyko')

html=BeautifulSoup(doc1.text, "html.parser")

skelbimai = html.select("a.classified")

for skelbimas in skelbimai:
        # 1 Pavadinimas
        pavadinimas_tag = skelbimas.select_one("h2.sl")
        pavadinimas = pavadinimas_tag.get_text(strip=True) if pavadinimas_tag else "Nežinomas pavadinimas"

        # 2 Nuoroda
        nuoroda = skelbimas.attrs.get("href", "#")
        pilna_nuoroda = f"https://www.skelbimai.lt{nuoroda}"

        # 3 Aprašymas
        aprasymas_tag = skelbimas.select_one("div.description")
        aprasymas = aprasymas_tag.get_text(strip=True) if aprasymas_tag else "Nėra aprašymo"

        # 4 Įkėlimo data
        ikelimo_data_tag = skelbimas.select_one("div.modified.f")
        ikelimo_data = ikelimo_data_tag.get_text(strip=True).replace("Atnaujinta:", "").strip() if ikelimo_data_tag else "Nežinoma data"

        # 5 Miestas
        miestas_tag = skelbimas.select_one("span.place")
        miestas = miestas_tag.get_text(strip=True) if miestas_tag else "Nežinomas miestas"

        # 6 Kaina
        kaina_tag = skelbimas.select_one("span.price-wp")
        kaina = kaina_tag.get_text(strip=True) if kaina_tag else "Kaina nenurodyta"

        #7 nuotrauka

        lb_div = skelbimas.select_one("div.lb")
        nuotraukos_adresas = "Nėra nuotraukos"
        if lb_div and "background-image" in lb_div.attrs.get("style", ""):
            style = lb_div.attrs["style"]
            nuotraukos_adresas = style.split("url(")[-1].split(")")[0]

        # Atspausdiname informaciją apie skelbimą
        print(f"Pavadinimas: {pavadinimas}")
        print(f"Nuoroda: {pilna_nuoroda}")
        print(f"Aprašymas: {aprasymas}")
        print(f"Įkėlimo data: {ikelimo_data}")
        print(f"Miestas: {miestas}")
        print(f"Kaina: {kaina}")
        print(f"Nuotrauka: {nuotraukos_adresas}")
        print("-" * 80)
```

```python
import requests
from bs4 import BeautifulSoup
import time  # importuojame time biblioteką

# Funkcija, kuri ištraukia informaciją iš puslapio
def get_page_data(url):
    doc1 = requests.get(url)

    if doc1.status_code != 200:
        print(f'Prisijungimas nepavyko prie {url}')
        return []

    html = BeautifulSoup(doc1.text, "html.parser")
    titles = html.select("a.classified")
    page_data = []

    # Pereiname per visus skelbimus
    for title in titles:
        pavadinimas_tag = title.select_one("h2.sl")
        pavadinimas = pavadinimas_tag.text.strip() if pavadinimas_tag else "Nežinomas pavadinimas"

        aprasymas_tag = title.select_one("div.description")
        aprasymas = aprasymas_tag.text.strip() if aprasymas_tag else "Nėra aprašymo"

        ikelimo_data_tag = title.select_one("div.modified")
        ikelimo_data = ikelimo_data_tag.text.replace('Atnaujinta:', "").strip() if ikelimo_data_tag else "Nėra datos"

        miestas_tag = title.select_one("span.place")
        miestas = miestas_tag.text.strip() if miestas_tag else "Nežinomas miestas"

        kaina_tag = title.select_one("span.price")
        kaina = kaina_tag.text.strip() if kaina_tag else "Kaina nenurodyta"

        nuoroda = title.attrs.get("href", "#")
        pilna_nuoroda = f"https://www.skelbimai.lt{nuoroda}/"

        lb_div = title.select_one("div.lb")
        nuotraukos_adresas = "Nėra nuotraukos"
        if lb_div and "background-image" in lb_div.attrs.get("style", ""):
            style = lb_div.attrs["style"]
            nuotraukos_adresas = style.split("url(")[-1].split(")")[0]

        page_data.append({
            "Pavadinimas": pavadinimas,
            "Nuoroda": pilna_nuoroda,
            "Aprašymas": aprasymas,
            "Įkėlimo data": ikelimo_data,
            "Miestas": miestas,
            "Kaina": kaina,
            "Nuotraukos adresas": nuotraukos_adresas
        })
    return page_data

# Funkcija, kuri pereina per visus puslapius
def scrape_all_pages(start_page, max_pages):
    base_url = "https://www.skelbimai.lt/paieska?category=74&page="
    all_data = []

    for page_num in range(start_page, max_pages + 1):
        url = f"{base_url}{page_num}"
        print(f"Renkuosi duomenis iš puslapio: {url}")
        page_data = get_page_data(url)

        if not page_data:
            print(f"Nėra daugiau skelbimų puslapyje {page_num}")
            break

        all_data.extend(page_data)

        # Vėlavimas tarp užklausų (2 sekundės)
        print(f"Vėluojama 2 sekundes prieš kitą puslapį...\n")
        time.sleep(2)  # Vėlavimas 2 sekundės (galite keisti pagal poreikį)

    return all_data

# Pavyzdinis naudojimas (rinkti duomenis nuo pirmo iki penkto puslapio)
data = scrape_all_pages(1, 10)

# Spausdiname gautą informaciją
for entry in data:
    print(f"Pavadinimas: {entry['Pavadinimas']}")
    print(f"Nuoroda: {entry['Nuoroda']}")
    print(f"Aprašymas: {entry['Aprašymas']}")
    print(f"Įkėlimo data: {entry['Įkėlimo data']}")
    print(f"Miestas: {entry['Miestas']}")
    print(f"Kaina: {entry['Kaina']}")
    print(f"Nuotraukos adresas: {entry['Nuotraukos adresas']}")
    print("-" * 50)
```
