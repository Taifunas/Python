```python
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Paleidžiame naršyklę
driver = webdriver.Chrome()
driver.maximize_window()

# Atidarome pagrindinį puslapį
driver.get("https://lytagra.lt")

# Priimame slapukus
try:
    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CLASS_NAME, "cookiesplus-accept-label"))).click()
    print("Slapukai priimti.")
except:
    print("Slapukų mygtukas nerastas arba jau priimtas.")

# Paspaudimai per meniu
try:
    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.LINK_TEXT, "Įrankiai ir įranga"))).click()
    print("Paspausta: 'Įrankiai ir įranga'.")
except:
    print("Nepavyko paspausti 'Įrankiai ir įranga'.")

try:
    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.LINK_TEXT, "Elektriniai įrankiai"))).click()
    print("Paspausta: 'Elektriniai įrankiai'.")
except:
    print("Nepavyko paspausti 'Elektriniai įrankiai'.")

try:
    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.LINK_TEXT, "Elektriniai kampiniai šlifuokliai"))).click()
    print("Paspausta: 'Elektriniai kampiniai šlifuokliai'.")
except:
    print("Nepavyko paspausti 'Elektriniai kampiniai šlifuokliai'.")

print("Atidarytas produktų sąrašas.")

# Išsaugome produktų sąrašo puslapio URL
products_page_url = driver.current_url

# Surandame visas produktų nuorodas ir limituojame iki 10 produktų
product_links = driver.find_elements(By.CSS_SELECTOR, ".thumbnail-container a")
product_urls = list(set([link.get_attribute("href") for link in product_links if link.get_attribute("href")]))[:10]

print(f"Rasti {len(product_urls)} produktai. Pradedame naršymą.")

# Saugojamas produktų sąrašas
product_data = []

# Einame per pirmus 10 produktų
for index, url in enumerate(product_urls):
    try:
        driver.get(url)
        print(f"Atidarytas produktas {index + 1}/{len(product_urls)}: {url}")

        # Laukiame, kol produktas užsikraus
        WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.CLASS_NAME, "product-essential")))

        # Ištraukiame produkto pavadinimą
        try:
            product_name = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CLASS_NAME, "product-name"))
            ).text
        except:
            product_name = "Nerastas"
            print(f"Klaida: Nepavyko gauti produkto pavadinimo ({url})")

        # Bandom ieškoti kainos dviejose galimose vietose
        product_price = "Nerasta"
        try:
            price_element = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, ".current-price .price[itemprop='price']"))
            )
            product_price = price_element.text
        except:
            try:
                # Jei nepavyko rasti "current-price", bandome alternatyvią vietą
                price_element = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.CSS_SELECTOR, ".price[itemprop='price']"))
                )
                product_price = price_element.text
            except:
                print(f"Klaida: Kaina nerasta ({url})")

        # Išsaugome produkto duomenis į sąrašą
        product_data.append({"Pavadinimas": product_name, "Kaina": product_price})
        print(f"Pavadinimas: {product_name}, Kaina: {product_price}")

        # Grįžtame į produktų sąrašą per URL
        driver.get(products_page_url)
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "thumbnail-container")))

    except Exception as e:
        print(f"Klaida lankantis puslapyje {url}: {str(e)}")

print("Produktų naršymas baigtas.")

# Išsaugome į CSV
csv_filename = "produktai.csv"
with open(csv_filename, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.DictWriter(file, fieldnames=["Pavadinimas", "Kaina"])
    writer.writeheader()
    writer.writerows(product_data)

print(f"Duomenys išsaugoti į {csv_filename}")

# Paliekame naršyklę atidarytą arba išjungiame ją
# driver.quit()
```
