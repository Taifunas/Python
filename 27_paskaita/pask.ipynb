```python
from selenium import webdriver
from time import sleep
from bs4 import BeautifulSoup
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import csv

# Initialize the Chrome driver
driver = webdriver.Chrome()

# Open the URL
driver.get("https://www.rimi.lt/e-parduotuve")
driver.maximize_window()

# Wait for the page to load
sleep(3)

# Decline cookies
mygtukas = driver.find_element(By.ID, "CybotCookiebotDialogBodyButtonDecline")
mygtukas.click()
sleep(2)

# Close the advertisement
reklama = driver.find_element(By.CSS_SELECTOR, ".modal__close")
reklama.click()
sleep(2)

# Search for mandarins
paieska = driver.find_element(By.ID, "search-input")
paieska.send_keys("mandarinai")
ieskoti = driver.find_element(By.CLASS_NAME, "form-field__search")
ieskoti.click()
sleep(2)

# Get product names
product_elements = driver.find_elements(By.CLASS_NAME, "card__url")
product_names = [product.get_attribute("aria-label") for product in product_elements]

# Print product names
for name in product_names:
    print(name)

# Save product names to a CSV file
with open("produktai.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Produkto pavadinimas"])  # Header
    for name in product_names:
        writer.writerow([name])

print("Duomenys išsaugoti į 'produktai.csv'!")

# Close the browser
driver.quit()
```
