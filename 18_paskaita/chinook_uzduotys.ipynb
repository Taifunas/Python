## Užduotis SELECT

- Išveskite visus duomenis iš lentelės invoices.
- Išveskite Billingcity ir BillingAddress iš lentelės invoices.
- Išveskite firstname and lastname iš lentelės customers, surikiuotas pagal pavardę, mažėjimo tvarka.
- Išveskite klientų id, vardus ir pavardes iš lentelės actor, surikiuotus pagal vardus didėjimo, pavardes mažėjimo tvarka

```python
import sqlite3
import pandas as pd


# Prisijungiame prie SQLite duomenų bazės
conn = sqlite3.connect('chinook.db')
cursor = conn.cursor()

print(' 1. Išvedame visus duomenis iš lentelės invoices')
cursor.execute("SELECT * FROM invoices")
invoices_data = cursor.fetchall()
invoices_df = pd.DataFrame(invoices_data, columns=[desc[0] for desc in cursor.description])
print("Visi duomenys iš invoices:")
#print(invoices_df)

print(' 2. Išvedame BillingCity ir BillingAddress iš lentelės invoices')
cursor.execute("SELECT BillingCity, BillingAddress FROM invoices")
billing_data = cursor.fetchall()
billing_df = pd.DataFrame(billing_data, columns=[desc[0] for desc in cursor.description])
print("BillingCity ir BillingAddress:")
print(billing_df)

print(' 3. Išvedame FirstName ir LastName iš lentelės customers, surikiuotus pagal pavardę mažėjimo tvarka')
cursor.execute("SELECT FirstName, LastName FROM customers ORDER BY LastName DESC")
customers_data = cursor.fetchall()
customers_df = pd.DataFrame(customers_data, columns=[desc[0] for desc in cursor.description])
print("Customers pagal pavardę mažėjimo tvarka:")
print(customers_df)

print(' 4. Išvedame artistų ID ir pavadinimus iš lentelės artists, surikiuotus pagal pavadinimą didėjimo tvarka')
cursor.execute("SELECT ArtistId, Name FROM artists ORDER BY Name ASC")
artists_data = cursor.fetchall()
artists_df = pd.DataFrame(artists_data, columns=["ArtistId", "Name"])
print("Artists pagal pavadinimą didėjimo tvarka:")
print(artists_df)

# Uždarome prisijungimą prie DB
conn.close()
```

## Užduotis WHERE

- Suraskite visus klientus, kurie yra iš Brazilijos.
- Suraskite visus užsakymus, kurie buvo atlikti 2010 metais.
- Suraskite vidutinę dainos, kurios žanras yra 2, kainą.

```python
import sqlite3
import pandas as pd


# Prisijungiame prie SQLite duomenų bazės
conn = sqlite3.connect('chinook.db')
cursor = conn.cursor()

# 5. Surandame visus klientus iš Brazilijos
cursor.execute("SELECT * FROM customers WHERE Country = 'Brazil'")
brazil_customers = cursor.fetchall()
brazil_customers_df = pd.DataFrame(brazil_customers, columns=[desc[0] for desc in cursor.description])
print("Klientai iš Brazilijos:")
print(brazil_customers_df)

# 6. Surandame visus užsakymus, atliktus 2010 metais
cursor.execute("SELECT * FROM invoices WHERE strftime('%Y', InvoiceDate) = '2010'")
invoices_2010 = cursor.fetchall()
invoices_2010_df = pd.DataFrame(invoices_2010, columns=[desc[0] for desc in cursor.description])
print("Užsakymai atlikti 2010 metais:")
print(invoices_2010_df)

# 7. Surandame vidutinę dainos, kurios žanras yra 2, kainą
cursor.execute("SELECT AVG(UnitPrice) FROM tracks WHERE GenreId = 2")
avg_price = cursor.fetchone()[0]
print("Vidutinė dainos kaina žanre 2:", avg_price)

# Uždarome prisijungimą prie DB
conn.close()
```

## Užduotis Funkcijos

- Pateikite pirmus dešimt darbuotojų vardų ir pavardžių pateikiant atsakymą didžiosiomis raidėmis.
- Pakeiskite visas a raides tekstu "testas" lentelėje employees, stulpelyje firstname.
- Parodykite visų dainų pavadinimų ilgius.
- Suskaičiuokite kiek sąskaitų turime (count).

```python
import sqlite3
import pandas as pd

# Prisijungiame prie SQLite duomenų bazės
conn = sqlite3.connect('chinook.db')
cursor = conn.cursor()

# 8. Pateikiame pirmus dešimt darbuotojų vardų ir pavardžių didžiosiomis raidėmis
cursor.execute("SELECT UPPER(FirstName), UPPER(LastName) FROM employees LIMIT 10")
employees_data = cursor.fetchall()
employees_df = pd.DataFrame(employees_data, columns=["FirstName", "LastName"])
print("Pirmi dešimt darbuotojų didžiosiomis raidėmis:")
print(employees_df)

# 9. Pakeičiame visas 'a' raides tekstu "testas" lentelėje employees, stulpelyje firstname
cursor.execute("SELECT REPLACE(FirstName, 'a', 'testas') FROM employees")
modified_names = cursor.fetchall()
modified_names_df = pd.DataFrame(modified_names, columns=["ModifiedFirstName"])
print("Pakeisti vardai su 'testas':")
print(modified_names_df)

# 10. Parodome visų dainų pavadinimų ilgius
cursor.execute("SELECT Name, LENGTH(Name) FROM tracks")
track_lengths = cursor.fetchall()
track_lengths_df = pd.DataFrame(track_lengths, columns=["TrackName", "Length"])
print("Dainų pavadinimų ilgiai:")
print(track_lengths_df)

# 11. Suskaičiuojame kiek sąskaitų turime
cursor.execute("SELECT COUNT(*) FROM invoices")
invoice_count = cursor.fetchone()[0]
print("Bendras sąskaitų skaičius:", invoice_count)

# Uždarome prisijungimą prie DB
conn.close()
```
