```python
import sqlite3

with sqlite3.connect('chinook.db') as conn:
    c = conn.cursor()
    tracks = c.execute("""SELECT 
                       tracks.name, 
                       unitprice,
                       genres.name
                       FROM tracks 
                       JOIN genres 
                       ON tracks.genreid = genres.genreid""").fetchall()

for track in tracks:
    print(track)
```

```python
import sqlite3

with sqlite3.connect('chinook.db') as conn:
    c = conn.cursor()
    tracks = c.execute("""SELECT
                       unitprice,
                       genreid,
                       COUNT(unitprice)
                       FROM tracks
                       GROUP BY unitprice, genreid
                       """).fetchall()

for track in tracks:
    print(track)
```

```python
import sqlite3

with sqlite3.connect('chinook.db') as conn:
    c = conn.cursor()
    tracks = c.execute("""SELECT 
                       tracks.name, 
                       unitprice,
                       genres.name
                       FROM tracks 
                       JOIN genres 
                       ON tracks.genreid = genres.genreid
                       ORDER BY unitprice DESC""").fetchall()

for track in tracks:
    print(track)
```

```python
import sqlite3

with sqlite3.connect('chinook.db') as conn:
    c = conn.cursor()
    len_genres = len(c.execute(""" SELECT * FROM genres""").fetchall())
    gen_ids = c.execute(""" SELECT * FROM genres""").fetchall()
    for gen_id in gen_ids:
        id = gen_id[0]
        print("Genre type: ", gen_id[1])
        tracks = c.execute(f"""SELECT 
                       name
                       FROM tracks
                       where genreid = {id} """).fetchall()
        names = " | ".join([track[0] for track in tracks])
        print(names)
        print("----------------------")
```
